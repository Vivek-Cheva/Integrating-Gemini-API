```python
import speech_recognition as sr

# Initialize recognizer
recognizer = sr.Recognizer()

def listen_and_convert():
  """
  This function listens to microphone input and converts it to text.
  """
  with sr.Microphone() as source:
    print("Listening...")

    # Adjust for ambient noise for better accuracy
    recognizer.adjust_for_ambient_noise(source, duration=0.5)

    # Listen for audio
    audio = recognizer.listen(source)

  try:
    print("Recognizing...")
    text = recognizer.recognize_google(audio)
    print("You: " + text)
    return text

    
  except sr.UnknownValueError:
    return "Sorry, could not understand audio"
  except sr.RequestError as e:
    return "Could not request results from Google Speech Recognition service"

# Run the function to listen and convert
# listen_and_convert()
```

```python
import pyttsx3

# Initialize the text-to-speech engine
engine = pyttsx3.init()

def speak(text):
  """
  This function takes text as input and speaks it aloud.
  """
  engine.say(text)
  engine.runAndWait()

# Example usage
text_to_speak = "This text will be spoken in real-time."
speak(text_to_speak)

print("Text spoken!")
```
